```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
path = "./wholesome.xlsx"
df = pd.read_excel(path,"source of funds")
df = df.replace(to_replace=".",value=0)
df.head()
```

```python
df.shape
```

```python
sector_total = df.iloc[0:34,2:].values
sector_total = sector_total / sector_total[0]
sector_total = pd.DataFrame(sector_total)
sector_total = pd.concat([df["Unnamed: 1"][0:34],sector_total],axis=1)
sector_total = sector_total.iloc[1:]
sector_total.head()
```

```python
average = sector_total.iloc[:,1:].sum(axis=1)/21.0
sector_total["Average"] = average
sector_total.head()
```

```python
sector_total.sort_values(by=["Average"], axis=0, inplace=True)
sector_total
```

```python
relevant_fundings = sector_total.iloc[::-1].iloc[:5,:]
relevant_fundings_index = relevant_fundings.index.values
relevant_fundings_index = np.append([0],relevant_fundings_index)
print(relevant_fundings_index)
```

```python
relevant_fundings = relevant_fundings.reset_index().drop(["index"],axis=1)
```

```python
relevant_funds = relevant_fundings["Unnamed: 1"].values
relevant_funds
```

```python
arr = []
for i in range(19):
    indexes = relevant_fundings_index + i*34
    arr = np.append(arr,indexes)
```

```python
filtered_df = df.iloc[arr]
filtered_df = filtered_df.reset_index().drop(["index"],axis=1)
print(filtered_df.shape)
filtered_df.head(15)
```

```python
   
totals = [6*i for i in range(19)]
average = filtered_df.iloc[totals,2:].sum(axis=1)/21.0
relevant_sectors = average.sort_values().index[-6:]
relevant_sectors
```

```python
arr = []
for i in range(6):
    arr = np.append(arr,relevant_sectors+i)
arr = np.sort(arr)
print(arr)
```

```python
filtered_df = filtered_df.iloc[arr]
filtered_df = filtered_df.reset_index().drop(["index"],axis=1)
print(filtered_df.shape)
(filtered_df)
```

```python
pd.unique(filtered_df["Unnamed: 0"])
```

```python
pd.unique(filtered_df["Unnamed: 1"])
```

```python
file_name = "rnd_sector.xlsx"
filtered_df.to_excel(file_name)
```

```python
filtered_df.to_excel("data.xlsx",sheet_name="Sector by Funding")
```

```python
df = pd.read_excel("./wholesome.xlsx","FDI by industry")
# df = df.iloc[[0,4]]
# df = df.reset_index().drop(["index"],axis=1)
df = df.set_index(df["Unnamed: 0"].values)
df = df.drop(["Unnamed: 0"], axis=1)
df
```

```python
ax = plt.subplot()
fig = plt.figure(figsize=(20,15))
df = (df-df.mean())/df.std()
sns.heatmap(df, cmap="ocean", annot=True, fmt=".2f")

plt.title("Normalized Sectoral FDI ")
plt.ylabel("Sectors")
plt.xlabel("Years")
plt.savefig("./plots/sectoral fdi heatmap", dpi=600, facecolor='w', edgecolor='w', orientation='portrait', format=None, transparent=False, bbox_inches=None, pad_inches=0.1, metadata=None)

plt.show()
```
